```python
import io, whisper
import numpy as np
from pydub import AudioSegment
from tqdm.auto import tqdm   # pip install tqdm

audio_path =r"Z:/03_my_program/音频处理/pythonProject1/20250901_174606_1.m4a"
SEGMENT_MS = 15 * 60 * 1000   # 每段 15 分钟
model = whisper.load_model("base")

# 读取整个音频
audio = AudioSegment.from_file(audio_path)

for idx, start_ms in enumerate(range(0, len(audio), SEGMENT_MS), 1):
    end_ms = min(start_ms + SEGMENT_MS, len(audio))
    segment = audio[start_ms:end_ms]

    # 转 numpy array（float32，单声道，归一化到 -1~1）
    samples = np.array(segment.get_array_of_samples()).astype(np.float32)
    if segment.channels > 1:
        samples = samples.reshape((-1, segment.channels))
        samples = samples.mean(axis=1)  # 转单声道
    samples /= np.iinfo(segment.array_type).max

    duration = len(segment) / 1000.0
    print(f"\n===== 第 {idx} 段 ({start_ms//1000}s – {end_ms//1000}s) =====")

    # tqdm 进度条（这里只能模拟，Whisper 没有原生回调）
    with tqdm(total=round(duration, 2), unit="s") as bar:
        result = model.transcribe(
            samples,
            language="zh",
            verbose=None,
            word_timestamps=False,
        )
        bar.update(bar.total)   # 直接拉满进度条

    print(result["text"].strip())
```
